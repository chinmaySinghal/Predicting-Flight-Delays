```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
cwd = os.getcwd()
datadir = '/'.join(cwd.split('/')[0:-1]) + '/Data/BTS/'

allFiles = glob.glob(datadir + "*.csv")
frame = pd.DataFrame()
list_ = []
count=0
for file_ in allFiles:
    count+=1
    df = pd.read_csv(file_)
    list_.append(df)
frame = pd.concat(list_,sort=False)
frame=frame[frame['ORIGIN']=='JFK']
frame=frame.drop(['Unnamed: 7','ORIGIN'],axis=1)
frame.head()
```

```python
frame.to_csv('2013Aug-2016Jul.csv')
```

```python
df1=pd.read_csv('2016Aug-2017Jul.csv')
df2=pd.read_csv('2017Aug-2018Jul.csv')
df=pd.concat([df1,df2])
df=df.drop(['Unnamed: 0','TAIL_NUM','ORIGIN_AIRPORT_ID','ORIGIN_CITY_MARKET_ID','Unnamed: 9'],axis=1)
data=pd.concat([df,frame])
data.head()
```

```python
data['MONTH'] = data['MONTH'].apply(lambda x: '{0:0>2}'.format(x))
data['DAY_OF_MONTH'] = data['DAY_OF_MONTH'].apply(lambda x: '{0:0>2}'.format(x))
data["DATE"] = data['YEAR'].map(str)+'-'+data["MONTH"].map(str) +'-'+ data['DAY_OF_MONTH'].map(str)
data.head()
```

```python
weather=pd.read_csv("Updated_weather.csv")
merged_df=data.merge(weather,on='DATE')
merged_df.head()
```

```python
merged_df.to_csv('Weather_BTS.csv')
```
